1. Find all trips that is greater than 10 miles and originated from New York and Morris

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("My Uber Drives - 2016.csv")
data
```

```python
x = data[data["START*"].isin(["New York","Morrisville"])]
```

```python
x.loc[x["MILES*"]>10, ["START*", "MILES*"]]
```

2. Create a new column with the following condition:
   a) >10 : Long Trip
   b) 5-10 : Medium Trip
   c) <5 : Short Trip

a) >10 : Long Trip

```python
data["new_column"] = np.where(data["MILES*"] > 10, "Long Trip", "Medium Trip")
```

```python
data.loc[data["MILES*"]>10, ["MILES*", "new_column"]]
```

b) 5-10 : Medium Trip

```python
#data["new_column"] = np.where((data["MILES*"] >= 5) & (data["MILES*"] <= 10), ["Medium Trip"], ["Medium Trip"])
```

```python
data.loc[(data["MILES*"] >= 5) & (data["MILES*"] <= 10), ["MILES*", "new_column"]]
```

c) <5 : Short Trip

```python
data["new_column"] = np.where(data["MILES*"] < 5, "Short Trip", "Medium Trip")
```

```python
data.loc[data["MILES*"]<5, ["MILES*", "new_column"]]
```
