```python
#import model yang dibutuhkan
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Membaca file csv
df = pd.read_csv('hotel.csv')
df
```

```python
#properties data csv
print(df.info())

#Jumlah data
count_null = df.isna().sum().sum()
count_notnull = df.notnull().sum().sum()
print('jumlah data mising values : ', count_null)
print()
print('list kolom dengan data mising values : ')
print(df.isna().sum())
print()
print('jumlah data dengan values : ', count_notnull)
print('jumlah seluruh data : ', (count_null + count_null))

#grafik dtypes pada data
slices = [4, 16, 12]
labels = ['float', 'int', 'object']

plt.pie(slices, labels=labels, autopct = '%1.1f%%')

plt.title("Data Types")
plt.tight_layout()
plt.show()

#grafik mising values data
slices = [488, 16340, 4, 112593]
labels = ['country', 'agent', 'children', 'company']

plt.pie(slices, labels=labels, autopct = '%1.1f%%')

plt.title("Mising Values")
plt.tight_layout()
plt.show()
```

```python
#count, mean, std, dan median data
df_t = df.describe().T
df_t2 = df_t.loc[:,['count','mean','std','50%']]
df_t2.columns = ['Jumlah Data', 'Mean', 'Std', 'Median']
df_t2
```

```python
#min dan max
df_t = df.describe().T
df_t[['min', 'max']]
```

```python
#drop kolom dengan 50% lebih mising value
#hitung jumlah theshold mising values (50%)
persen = int(0.5*df.shape[0] + 1)

#cetak grafik kolom vs jumlah mising values
jumlah = [488, 16340, 4, 112593]
kolom = ['country', 'agent', 'children', 'company']

miss_val = pd.DataFrame({"Columns":kolom, "Jumlah":jumlah})
miss_val = miss_val.sort_values('Jumlah')
plt.bar('Columns', 'Jumlah', data = miss_val)

plt.title("Columns with mising Values")
plt.axhline(y=persen, linestyle = '--', c = 'r', label = 'threshold 50% + 1')
plt.xlabel("Columns")
plt.ylabel("Number of mising Values")
plt.legend()

plt.tight_layout()
plt.show()

#menghapus kolom yang melebihi treshold.
df.dropna(axis = 1, thresh = persen, inplace = True)
df.info()
```

```python
#mengisi mising value dengan modus

#mengecek modus dari kolom
print('modus dari country :', df['country'].mode()[0])
print('modus dari children :', df['children'].mode()[0])
print()

#mengisi kolom tersebut dengan modus
df['country'] = df['country'].fillna(df['country']. mode()[0])
df['children'] = df['children'].fillna(df['children']. mode()[0])
print(df.isna().sum())
print()

#cetak grafik children
slices = df['children'].value_counts()
labels = df['children'].value_counts().index.tolist()
myexplode = [0, 0, 0, 0.5, 1.5]

plt.pie(slices, labels=labels, explode = myexplode, autopct = '%1.1f%%')

plt.title("Children Graph")
plt.tight_layout()
plt.show()
```

```python
#cetak Grafik Country
print('5 data negara terbanyak')
jumlah = df['country'].value_counts().head()
country = df['country'].value_counts().head().index.tolist()

miss_val = pd.DataFrame({"Columns":country, "Jumlah":jumlah})
miss_val = miss_val.sort_values('Jumlah', ascending=False)
plt.bar('Columns', 'Jumlah', data = miss_val)

plt.title("Country Graph")
plt.xlabel("country")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

```python
#Missing value denganmean
#mencari mean dari kolom agent
print('mean dari agent :', df['agent'].mean())
print()

#mengisi kolom tersebut dengan mean untuk missing value
df['agent'] = df['agent'].fillna(df['agent']. mode())
print(df.isna().sum())
print()

#cetak grafik agent
print('kolom agent 5 data terbanyak')
slices = df['agent'].value_counts().head()
labels = df['agent'].value_counts().head().index.tolist()
myexplode = [0, 0, 0, 0, 0]

plt.pie(slices, labels=labels, explode = myexplode, autopct = '%1.1f%%')

plt.title("Agent Graph")
plt.tight_layout()
plt.show()
```

```python
#cetak Grafik agent
print('5 data agent terbanyak')
jumlah = df['agent'].value_counts().head()
agent = df['agent'].value_counts().head().index.tolist()

miss_val = pd.DataFrame({"Columns":agent, "Jumlah":jumlah})
miss_val = miss_val.sort_values('Jumlah', ascending=False)
plt.bar('Columns', 'Jumlah', data = miss_val)

plt.title("Agent Graph")
plt.xlabel("Agent")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

```python
#Mengkategorikan lead time
condition = [
    (df['lead_time'] < 30),
    (df['lead_time'] >= 30) & (df['lead_time'] <= 80),
    (df['lead_time'] > 80)
    ]
#Output
values = ['Short', 'Medium', 'Long']
#tambah kolom isi data kondisi
df['Duration'] = np.select(condition, values)

#cetak grafik duration
slices = df['Duration'].value_counts().head()
labels = df['Duration'].value_counts().head().index.tolist()

plt.pie(slices, labels=labels, autopct = '%1.1f%%')

plt.title("Duration")
plt.tight_layout()
plt.show()
df['Duration']
```
